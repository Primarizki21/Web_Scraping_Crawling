### Library

```python
import requests
import json
import pandas as pd
```

### Response API

#### Response menggunakan Post dan Simpan sebagai JSON data

```python
response = requests.post(url_target, headers=headers, data=query)
products = response.json()[0]['data']['GetShopProduct']['data']
```

Maksud dari [0]['data']['GetShopProduct']['data'] adalah 
- [0] untuk data mengakses data pertama dari JSON
- ['data'] untuk mengakses key 'data'
- ['GetShopProduct'] untuk mengakses 'GetShopProduct' pada 'data'
- ['data'] untuk mengakses data dari 'GetShopProduct'

#### Pretty Print JSON Data

```python
json_data = response.json()
print(json.dumps(json_data, indent=2))
```

### Dataframe Barang

Kita akan menyimpan data-data berikut:
- Nama Produk
- Link gambar
- Harga Produk (harga sebelum diskon dan setelah diskon)
- Diskon Produk
- Jumlah Review
- Rata-rata Nilai Rating

```python
list_produk = []
for data in products:
    nama_produk = data['name']
    link_gambar = data['primary_image']['original']
    harga_produk_sebelum_disk = data['campaign']['original_price_fmt']
    harga_produk_sesudah_disk = data['price']['text_idr']
    diskon_produk = data['campaign']['discounted_percentage']
    jumlah_review = data['stats']['reviewCount']
    avrg_rating = data['stats']['averageRating']
    
    x = {
        'Nama': nama_produk,
        'Gambar': link_gambar,
        'Harga Sebelum Diskon': harga_produk_sebelum_disk,
        'Harga Sesudah Diskon': harga_produk_sesudah_disk,
        'Diskon': diskon_produk,
        'Jumlah Review': jumlah_review,
        'Rata-rata Rating': avrg_rating
    }
    
    list_produk.append(x)
```

```python
df_barang = pd.DataFrame(list_produk)
df_barang.to_csv('hasil scraping tokped anker indonesia.csv', index=False, encoding='utf-8')
df_barang.to_excel('hasil scraping tokped anker indonesia.xlsx', index=False)
```
